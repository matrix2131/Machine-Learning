**DIRECTLY DRAWING WITH MOUSE**

```python
def draw_circle(event,x,y,flags,param):  # The parameters are automatically taken by openCV.
    
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img,(x,y), 100, (255,0,0), -1)


cv2.namedWindow(winname ="MY DRAWING")

cv2.setMouseCallback("MY DRAWING",draw_circle)

# This is the way of joining the function with image being taken in.



#######################################
##### SHOWING IMAGE WITH OPENCV   #####
#######################################


img = np.zeros((512,512,3), np.int8)

while True:
    cv2.imshow("MY DRAWING", img)
    
    if cv2.waitKey(20) & 0xFF == 27:
        break
cv2.destroyAllWindows()


# Run this in terminal 
# The windows displayed will be gray in color because of the npint8 used.
# The window will be black if we remove np.int8.
```

```python
def draw_circle(event,x,y,flags,param):  # The parameters are automatically taken by openCV.
    
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img,(x,y), 100, (255,0,0), -1)
        
    elif event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img,(x,y), 100, (0,255,0), -1)


cv2.namedWindow(winname ="MY DRAWING")

cv2.setMouseCallback("MY DRAWING",draw_circle)

# This is the way of joining the function with image being taken in.



#######################################
##### SHOWING IMAGE WITH OPENCV   #####
#######################################


img = np.zeros((512,512,3), np.int8)

while True:
    cv2.imshow("MY DRAWING", img)
    
    if cv2.waitKey(20) & 0xFF == 27:
        break
cv2.destroyAllWindows()


# Run this in terminal 
# The windows displayed will be gray in color because of the npint8 used.
# The window will be black if we remove np.int8.
```
