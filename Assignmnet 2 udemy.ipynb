```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/home/soniya/Documents/work/Computer-Vision-with-Python/DATA/dog_backpack.jpg")
```

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img3 = img2.copy()
```

```python
plt.imshow(img2);
```

```python
flipped = cv2.flip(img2,0)
plt.imshow(flipped)
```

```python
face = cv2.rectangle(img2, (200,350), (600,750), (255,0,0), thickness = 5)
plt.imshow(face)
```

```python
vertices = np.array([[350,350],[200,700],[650,700]], dtype = np.int32)
```

```python
pts = vertices.reshape((-1,1,2))
```

```python
cv2.polylines(img3, [pts], isClosed = True, color = (0,0,255),thickness = 10)
```

```python
plt.imshow(img3)
```

```python
cv2.fillPoly(img3,[pts],(0,0,255))
```

```python
plt.imshow(img3)
```

```python
def draw_circle(event,x,y,flags,param):
    if event == cv2.EVENT_RBUTTONDOWM:
        cv2.circle(img,(x,y), 50, (255,0,0) , 8)
    
cv2.namedWindow(winname = "CIRCLES")
cv2.setMouseCallback("CIRCLES", draw_circle)

img = cv2.imread("/home/soniya/Documents/work/Computer-Vision-with-Python/DATA/dog_backpack.jpg")

while True:
    cv2.imshow("CIRCLES", img)
    
    if cv2.waitKey(20) & 0xFF == 27:
        break
        
cv2.destroyAllWindows()
```
