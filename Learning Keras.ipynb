Now we are going to separate labels from the actual features.

```python
labels = data[:,4]
```

```python
features = data[:,0:4]
```

```python
features
```

```python
X = features
y = labels
```

```python
from sklearn.model_selection import train_test_split
# This splits the data in two parts...that is train data and test data 
# and also distributes them randomly.
```

```python
np.__version__
```

## As the numpy's latest version has removed numpy.testing.nosetester.....so I have downgraded the version of numpy.

```python
# Test_size says that 33% will go to X_test.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# We scale or standardize the data.
from sklearn.preprocessing import MinMaxScaler
# This is used if the difference between the X_test.max() and X_test.min()
# is large.
```

```python
scalar_object = MinMaxScaler()
```

```python
scalar_object.fit(X_train)
```

```python
scaled_X_train = scalar_object.transform(X_train)
```

```python
scaled_X_test = scalar_object.transform(X_test)
```

```python
scaled_X_train.max()
```

```python
X_train
```

```python
scaled_X_train
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(4,input_dim = 4, activation = 'relu'))
# '4' is the number of neurons or features.This basically is the input layer.

model.add(Dense(8,activation = 'relu'))
# This is the hidden layer.

model.add(Dense(1, activation = 'sigmoid'))
```

```python
# Now we will compile the model.
model.compile(loss = 'binary_crossentropy', optimizer = 'adam',metrics = ['accuracy'])
```

```python
model.fit(scaled_X_train, y_train, epochs = 50, verbose = 2)
# One epoch is going through the data one time.
```

```python
# Predicting on new data.
model.predict_classes(scaled_X_test)
```

```python
# Now we will evaluate the model.
model.metrics_names
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
predictions = model.predict_classes(scaled_X_test)
```

```python
confusion_matrix(y_test,predictions)
# This will give us our confusion matrix.
```

```python
print(classification_report(y_test, predictions))
```
